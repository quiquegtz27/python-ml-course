# Variables dummy
Funciona para separar una columna **categorica** en el número de columnas como categorías exista en la columna original.
A este proceso se le llama "dummificación".


```python
data["sex"].head(10)
```

```python
dummy_sex = pd.get_dummies(data["sex"], prefix="sex")
```

```python
dummy_sex.head(10)
#"Dumificación"
```

```python
column_name=data.columns.values.tolist()
column_name
```

```python
data = data.drop(["sex"], axis = 1)
```

```python
data = pd.concat([data, dummy_sex], axis = 1)
```

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy ], axis = 1)
    return df
```
